Question 1:

What is a random variable in probability theory?
Answer:
A random variable is a function that assigns a numerical value to each outcome in a probability experiment. It maps outcomes of a random process to numbers, allowing mathematical and statistical analysis.

Question 2:

What are the types of random variables?
Answer:
There are two main types of random variables:

Discrete Random Variable: Takes countable values (e.g., number of heads in coin tosses).

Continuous Random Variable: Takes any value within an interval (e.g., height, weight).
Some sources also mention Mixed Random Variables, which have both discrete and continuous parts.

Question 3:

Explain the difference between discrete and continuous distributions.
Answer:

Discrete distributions assign probability to specific values (using a Probability Mass Function).
Example: The probability of getting exactly 2 heads in 3 coin tosses.

Continuous distributions assign probability density over intervals (using a Probability Density Function).
The probability of getting an exact value is zero; probabilities are calculated over ranges.

Question 4:

What is a binomial distribution, and how is it used in probability?
Answer:
A binomial distribution gives the probability of getting exactly k successes in n independent Bernoulli trials, each with probability p of success.
Formula:
   P(X=k) =C(n,k) p^k (1-p)^(n-k).
Uses: Coin tosses, quality control, pass/fail experiments.

Question 5:

What is the standard normal distribution, and why is it important?
Answer:
The standard normal distribution is a normal distribution with a mean of 0 and standard deviation of 1.
It is important because many statistics (like z-scores) can be transformed to this distribution, making it easier to compare and calculate probabilities.

Question 6:

What is the Central Limit Theorem (CLT), and why is it critical in statistics?
Answer:
The Central Limit Theorem states that the sampling distribution of the sample mean approaches a normal distribution as the sample size becomes large, regardless of the original population’s shape.
It’s critical because it allows us to use normal distribution methods (like confidence intervals and hypothesis testing) even when data are not perfectly normal.

Question 7:

What is the significance of confidence intervals in statistical analysis?
Answer:
A confidence interval gives a range of likely values for a population parameter.
For example, a 95% confidence interval means that if we repeat the sampling process many times, about 95% of such intervals will contain the true population mean.
It helps express the uncertainty in our estimates.

Question 8:

What is the concept of expected value in a probability distribution?
Answer:
The expected value (mean) of a random variable is the long-run average outcome.

For discrete random variables:

E[X]=∑x⋅P(X=x)

For continuous random variables:

E[X]=∫xf(x)dx

It represents the theoretical average if an experiment were repeated infinitely many times.

```python
#Write a Python program to generate 1000 random numbers from a normal distribution with mean = 50 and standard deviation = 5. Compute its mean and standard deviation using NumPy, and draw a histogram to visualize the distribution.
import numpy as np
import matplotlib.pyplot as plt


np.random.seed(42)
samples = np.random.normal(loc=50, scale=5, size=1000)


print("Mean =", samples.mean())
print("Standard deviation =", samples.std(ddof=1))

plt.hist(samples, bins=30)
plt.title("Histogram of 1000 samples from N(50, 5)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
'''You are working as a data analyst for a retail company. The company has collected daily sales data for 2 years and wants you to identify the overall sales trend.
daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
 235, 260, 245, 250, 225, 270, 265, 255, 250, 260]
● Explain how you would apply the Central Limit Theorem to estimate the average sales
with a 95% confidence interval.
● Write the Python code to compute the mean sales and its confidence interval.'''
import numpy as np

daily_sales = np.array([220,245,210,265,230,250,260,275,240,255,
                        235,260,245,250,225,270,265,255,250,260])

n = len(daily_sales)
mean = daily_sales.mean()
s = daily_sales.std(ddof=1)
se = s / np.sqrt(n)
z95 = 1.96
ci_lower = mean - z95 * se
ci_upper = mean + z95 * se

print("Mean sales =", mean)
print("95% Confidence Interval =", (ci_lower, ci_upper))
```
